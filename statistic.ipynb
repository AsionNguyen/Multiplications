```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "out_512_500.txt"
```

```python
data=pd.read_csv(file, sep=',')
```

```python
data
```

```python
X = data["Karatsuba"]
Y = data["SS"]
Z = data["Optimized"]
```

```python
plt.plot(X, color='r', label="Karatsuba")
plt.plot(Y, color='b', label="SS")
plt.plot(Z, color='g', label="Optimized")
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("samplers")
plt.ylabel("seconds")
plt.title("Flint Multiplication comparaison")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```

```python
plt.plot(X, 'ro', markersize=1, label="Karatsuba")
plt.plot(Y, 'bo', markersize=1, label="SS")
plt.plot(Z, 'go', markersize=1, label="Optimized")
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("samplers")
plt.ylabel("seconds")
plt.title("Flint Multiplication comparaison")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
